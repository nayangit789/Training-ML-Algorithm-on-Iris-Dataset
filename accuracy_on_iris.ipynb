```python
import pandas as pd



from sklearn.ensemble import RandomForestClassifier


data = pd.read_excel("iris-train.xlsx")
data.head()


x = data.iloc[:,:-1]
y= data.iloc[:,-1]
model= RandomForestClassifier()
model.fit(x,y)
predictions = model.predict([[5.3, 3.7,1.5,0.2]])
print(predictions)


datatest=pd.read_excel("iris-test.xlsx")
x_test= datatest.iloc[:,:-1]
y_test= datatest.iloc[:, -1] # Corrected: Select only the last column for labels
predictions = model.predict(x_test.values)
print (predictions)
print(y_test.values)
```

```python
actual = y_test.values
```

```python
count = 0
for i in range(0, len(actual)):
    if actual[i] == predictions[i]:
        count = count + 1
print((count* 100)/len(actual))
```
