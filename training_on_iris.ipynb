```python
import pandas as pd
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_excel("iris-test.xlsx")
data.head()
```

```python
x = data.iloc[:,:-1]
y= data.iloc[:,-1]
model= RandomForestClassifier()
model.fit(x,y)
predictions = model.predict([[5.3, 3.7,1.5,0.2]])
print(predictions)
```

```python
datatest=pd.read_excel("iris-test.xlsx")
x_test= datatest.iloc[:,:-1]
y_test= datatest.iloc[:-1]
predictions = model.predict(x_test.values)
print (predictions)
```
